```python
#label encoding 
Cancer_data.replace({'diagnosis':{'M':0,'B':1}}, inplace=True)
```

```python
#spliting the dataset in feature and target
X=Cancer_data.drop(['diagnosis','id','Unnamed: 32'],axis=1)
Y=Cancer_data['diagnosis']
```

```python
print(X)
print(Y)
```

```python
#spliting the data in train and test data
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,random_state=2, test_size=0.2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

# Train the model with LogisticRegressor

```python
model = LogisticRegression()
```

```python
#fitting the model in train and test dataset
model.fit(X_train,Y_train)
```

# Model Evaluation

```python
#accuracy score on training dataset
training_prdiction = model.predict(X_train)
```

```python
training_data_accuracy = accuracy_score(training_prdiction,Y_train)
```

```python
#printing the accuracy score of the dataset
print('Accuracy score of the dataset: ',training_data_accuracy)
```

```python
#accuracy score on test dataset
testdata_prdiction = model.predict(X_test)
```

```python
test_data_accuracy = accuracy_score(testdata_prdiction,Y_test)
```

```python
#printing the accuracy score of the dataset
print('Accuracy score of the dataset: ',test_data_accuracy)
```
